```python
import pandas as pd
```

```python
expert = pd.read_csv('/jupyter/analysis/expert_table.csv')
```

```python
xml_df = pd.read_csv('/jupyter/output/XMLProduct_DBID_2.csv')
xml_df.head()
```

```python
xml_df1 = pd.read_csv('/jupyter/output/XMLProduct_DBID.csv')
xml_df1.head()
```

```python
print (len(xml_df1), len(xml_df))
xml_df = xml_df.append(xml_df1)
print (len(xml_df1))
```

```python
xml_df = xml_df.drop_duplicates(subset=['Label_ID','Section'], keep='first')
```

```python
expert['Section'] = 'Indications'
```

```python
index= expert.label_id.str.contains('./DailyMedExtracter/prescription/')
expert.loc[index,'Section']=  'Contraindications'
```

```python
expert.loc[index,'label_id'] = expert.loc[index,'label_id'].str.replace('../DailyMedExtracter/prescription/temp_xml/','')
```

```python
expert =expert.rename(columns={'do_id':'DO_ID','drug_id':'DB_ID','label_id':'Label_ID'})
```

```python
expert.head()
```

```python
expert =expert[expert.expert_consensus!='No consensus']
len(expert)
```

```python
match = expert.merge(xml_df, on=['DB_ID','Label_ID','Section'])
```

```python
match.head()
```

```python
columns = [ 'DB_ID', 'DO_ID', 'disease_name', 'Label_ID', 'expert_consensus',
       'Section', 'Set_ID', 'Drug_Brand_Name', 'Active_ingredient', 'UNII_ID',
       'Text', 'WordCount', 'Drug_name']
match = match[columns]
```

```python
match.rename(columns={'Text':'Context','expert_consensus':'relation','disease_name':'disease'}, inplace=True)
```

```python
match.to_csv('/data/output/match_expert_data.csv', index=False)
```
