```python
#Dataset which has been selected to be added to the goldstandard dataset. These instances have been choosen to be crowdsourced and with enough consensus added to the training dataset
df2 = pd.read_csv("../data/output/AdditionalGoldStandard.csv")
df2.head()
```

```python
#Drops rows which do not contain a DBID or a UMLSID as these are essential for the DDP
print(len(df2))
df2.dropna(subset=['disease', 'drug'], inplace=True)
print(len(df2))
```

```python
#Binary choices
y = []
for i, r in df2.iterrows():
    
    if r["Named_Cats"] == "['indication_treatment']" or r["Named_Cats"] == "['indication_symptomatic_relief']":
        y.append("Positive") 
    elif r["Named_Cats"] == "['contraindication']":
        y.append("Negative")
    else:
        print(r["Named_Cats"])
        y.append("NotNeeded")
    
```

```python
#Multiclass
y = []
for i, r in df2.iterrows():
    
    if r["Named_Cats"] == "['indication_treatment']":
        y.append("ind_treat")
    elif r["Named_Cats"] == "['indication_symptomatic_relief']":
        y.append("ind_symp")
    elif r["Named_Cats"] == "['contraindication']":
        y.append("contra")
    elif r["Named_Cats"] == "['effect']":
        y.append("effect")
    else:
        print(r["Named_Cats"])
        y.append("NotNeeded")
    
```

```python
#This is for the contraindication test
y = "contra"
```

```python
#Add the multiclass field and the true statement that this data has indeed been labeled before
df2["Class"] = y
df2["labeled?"] = True
df2.head()
```

```python
df2[df2.Class == "contra"]
```

```python
df2.columns
```

```python
df2.rename(columns={"drug":"Drug", "drug_name":"DrugName", "disease":"Disease", "disease_name":"Disease_Name", "context":"Context"}, inplace=True)
```

```python
dfsmall = df2[['Drug','Disease','DrugName', 'Disease_Name', 'Context', 'Class', 'labeled?']]
dfsmall.drop_duplicates(inplace = True)
dfsmall.head()
```

```python
dfsmall = dfsmall[dfsmall.Class != "NotNeeded"]
```

```python
dfsmall.Context[300]
```

```python
#innermerge
df3 = df.merge(dfsmall, on=["Drug", "Disease"], how = "inner")
df3.head()
```

```python
df3[df3.Class == "Negative"]
```

```python
#Concatenate the Original Gold standard and the unanimous crowdsourcing classifications
frames = [df, dfsmall]
dfNewGold = pd.concat(frames)
print(len(dfNewGold))
```

```python
dfNewGold[dfNewGold.Class == "contra"]
```

```python
#Keep the ones on the "Bottom" of the concat since some of these may overwrite the Classification of the original gold standard (EG. Positive gold standard to negative gold standard)
print(len(dfNewGold))
dfNewGold.drop_duplicates(["Drug","Disease"],keep="last", inplace=True)
print(len(dfNewGold))
```

```python
print(len(df))
print(len(df.drop_duplicates()))
print(len(dfsmall))
print(len(dfsmall.drop_duplicates()))
print(len(dfNewGold))
print(len(dfNewGold.drop_duplicates()))
print("Gold standard data grew from:", str(len(df)), "to ", str(len(dfNewGold)), "so increased by: ", str(len(dfNewGold)-len(df)))
```
