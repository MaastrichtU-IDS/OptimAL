```python
#THIS IS NOT NEEDED! WE ARE NOT INTERESTED IN DRUGS WITH MULTIPLE ACTIVE INGREDIENTS


#Seperate active ingredients within the active ingredient and UNII ID
count = 0
c2 = 0
for index, row in DM.iterrows():
    count = count + 1

    
    activeI = row['Active_ingredient']
    activeI = activeI.split("|")
    unii = row['UNII_ID']
    unii = unii.split("|")
    

    for i in range(len(unii)):
        
        c2 = c2 + 1
        indications = []
        data = [row['Label_ID'],row['Drug_Brand_Name'], activeI[i], unii[i], row['Formatted_Text'],row['Text']]
        indications = indications.append(data)  
    
        if count == 1 and c2 ==1:
            newDM = pd.DataFrame(indications, columns=['Label_ID','Drug_Brand_Name', 'Active_ingredient', 'UNII_ID', 'Formatted_Text','Text'])
        else:
            newrow = pd.DataFrame(indications, columns=['Label_ID','Drug_Brand_Name', 'Active_ingredient', 'UNII_ID', 'Formatted_Text','Text'])
            newDM = newDM.append(newrow)
    
```
